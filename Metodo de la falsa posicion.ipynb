Sea f(x)= x^3 + 4^2 - 10, encontrar la raiz que se encuentra en el intervalo [2,0] con un error permisible porcentual de 0.1%

```
def f(x):
  (x**3) + (4*x**2) -10
```

```
def compute_xr(xl,xu):
  return xu-((f(xu)*(xl-xu))/(f(xl)-f(xu)))
```

```
xu = 2
xl = 0
```

```
xr = compute_xr(xl,xu)

f_xl = f(xl)
f_xr = f(xr)
f_xu = f(xu)

print("El valor de xl es: ", xl)
print("El valor de xu es: ", xu)
print("El valor de xr es: ", xr)



print("El valor de f(xl) es: ", f_xl)
print("El valor de f(xr) es: ", f_xr)
print("El valor de f(xu) es: ", f_xu)

i_izquierdo= f_xl * f_xr
print("El valor de f(xl) * f(xr): ", f_xl * f_xr)
i_derecho= f_xu * f_xr
print("El valor de f(xu) * f(xr): ", f_xu * f_xr)

error = abs ((xu-xl)/(xu+xl))*100

if (i_izquierdo > 0):
  print("Se igualara el valor de xl por xr.")
  xl=xr
if (i_derecho > 0):
  print("Se igualara el valor de xu por xr.")
  xu=xr


print("El error permisible es: ", error)


```
